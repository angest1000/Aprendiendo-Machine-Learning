```python
#Visualizar datos con seaborn
#Codigo extraido del siguiente tutorial:

#https://www.youtube.com/watch?v=2tObZDNGTJI&t=0s

#Utilizaremos un dataset descargado desde kaggle
#Necesitamos loggearnos en Kaggle para descargar el dataset
#Link: https://www.kaggle.com/gregorut/videogamesales

#importamos librer√≠as
import seaborn as sns
import pandas as pd

#importamos datos
datos= pd.read_csv('vgsales.csv')

nuevo=datos[["Genre","Global_Sales","JP_Sales","EU_Sales"]]

```
